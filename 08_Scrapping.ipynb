```python
# Libraries
##  Load Content

import requests
## To parse the page Content
from bs4 import BeautifulSoup
## ..
import pandas as pd
import numpy as np 
```

```python
# URL of the page
url = 'https://www.investing.com/dividends-calendar/'

# Load the page
page = requests.get(url, verify = False)
```

```python
# Transformar the request  to html
dc = BeautifulSoup(page .content, 'html.parser')
dc
```

```python
# Identify the table
table = dc.find_all('table')

#DC
table1 = table[1]
table1
```

```python
# Find rows

rows = table1.find_all('tr')
```

```python
# Conver

tmp = []

for i in rows[1:]:
  cells = i.find_all(['td', 'th'])
  cells_text = [cell.get_text(strip=True) for cell in cells]

  tmp.append(cells_text)

data = pd.DataFrame(tmp)
data.head()
```
