```python
# The Libraries
import pandas as pd
import numpy as np # just in case
import plotly.express as px
import plotly.graph_objects as go  # graphs other than scatter and bar or customize
```

```python
# Save the file path
p1 = '/content/Data.xlsx'

# Read file
df = pd.read_excel(p1)

# Check
df.info()
```

```python
df = df.set_index('Date', drop = False) # drop does not delete Date
df.head()
```

```python
# Base Layout
fig = go.Figure(
    layout = go.Layout(
        updatemenus = [dict(type = "buttons", direction = "right", x = 0.9, y = 1.16)],
        xaxis = dict(range = ['1960','2022'],
                     autorange = False, tickwidth = 2,
                     title_text = "Dates"),
        yaxis = dict(range = [0,3500000],
                     autorange = False, 
                     title_text = ""),
        title = "Imports of goods and services (US$ at current prices)",
        title_font_size = 30,
        title_x = 0.5
    )
)
```

```python
# Add traces 
init = 1 

fig.add_traces(
    go.Scatter(x = df.Date[:init],
               y = df.BOL[:init],
               name = 'BOL',
               visible = True,
               line = dict(color = 'purple'),
               mode = 'lines+markers'
               )
)

fig.add_traces(
    go.Scatter(x = df.Date[:init],
               y = df.CAN[:init],
               name = 'CAN',
               visible = True,
               line = dict(color = 'purple'),
               mode = 'lines+markers'
               )
)

fig.add_traces(
    go.Scatter(x = df.Date[:init],
               y = df.CHN[:init],
               name = 'CHN',
               visible = True,
               line = dict(color = 'purple'),
               mode = 'lines+markers'
               )
)

fig.add_traces(
    go.Scatter(x = df.Date[:init],
               y = df.COL[:init],
               name = 'COL',
               visible = True,
               line = dict(color = 'purple'),
               mode = 'lines+markers'
               )
)

fig.add_traces(
    go.Scatter(x = df.Date[:init],
               y = df.DEU[:init],
               name = 'DEU',
               visible = True,
               line = dict(color = 'purple'),
               mode = 'lines+markers'
               )
)

fig.add_traces(
    go.Scatter(x = df.Date[:init],
               y = df.FRA[:init],
               name = 'FRA',
               visible = True,
               line = dict(color = 'purple'),
               mode = 'lines+markers'
               )
)

fig.add_traces(
    go.Scatter(x = df.Date[:init],
               y = df.GBR[:init],
               name = 'GBR',
               visible = True,
               line = dict(color = 'purple'),
               mode = 'lines+markers'
               )
)

fig.add_traces(
    go.Scatter(x = df.Date[:init],
               y = df.ITA[:init],
               name = 'ITA',
               visible = True,
               line = dict(color = 'purple'),
               mode = 'lines+markers'
               )
)

fig.add_traces(
    go.Scatter(x = df.Date[:init],
               y = df.JPN[:init],
               name = 'JPN',
               visible = True,
               line = dict(color = 'purple'),
               mode = 'lines+markers'
               )
)

fig.add_traces(
    go.Scatter(x = df.Date[:init],
               y = df.KOR[:init],
               name = 'KOR',
               visible = True,
               line = dict(color = 'purple'),
               mode = 'lines+markers'
               )
)

fig.add_traces(
    go.Scatter(x = df.Date[:init],
               y = df.NLD[:init],
               name = 'NLD',
               visible = True,
               line = dict(color = 'purple'),
               mode = 'lines+markers'
               )
)

fig.add_traces(
    go.Scatter(x = df.Date[:init],
               y = df.RUS[:init],
               name = 'RUS',
               visible = True,
               line = dict(color = 'purple'),
               mode = 'lines+markers'
               )
)

fig.add_traces(
    go.Scatter(x = df.Date[:init],
               y = df.USA[:init],
               name = 'USA',
               visible = True,
               line = dict(color = 'purple'),
               mode = 'lines+markers'
               )
)
```

```python
# Frames
frames = [
    go.Frame(
        data = [
            go.Scatter(x = df.Date[:k], y = df.BOL[:k]),
            go.Scatter(x = df.Date[:k], y = df.CAN[:k]),
            go.Scatter(x = df.Date[:k], y = df.CHN[:k]),
            go.Scatter(x = df.Date[:k], y = df.COL[:k]),
            go.Scatter(x = df.Date[:k], y = df.DEU[:k]),
            go.Scatter(x = df.Date[:k], y = df.FRA[:k]),
            go.Scatter(x = df.Date[:k], y = df.GBR[:k]),
            go.Scatter(x = df.Date[:k], y = df.ITA[:k]),
            go.Scatter(x = df.Date[:k], y = df.JPN[:k]),
            go.Scatter(x = df.Date[:k], y = df.KOR[:k]),
            go.Scatter(x = df.Date[:k], y = df.NLD[:k]),
            go.Scatter(x = df.Date[:k], y = df.RUS[:k]),
            go.Scatter(x = df.Date[:k], y = df.USA[:k]),
        ]
    )
    for k in range(init, len(df)+1)
]
```

```python
# Animation
fig.update(frames = frames)
```

```python
# Buttons
fig.update_layout(
    updatemenus = [
        dict(
            buttons = list([
                dict(label = 'Play',
                     method = 'animate',
                     args = [None, {'frame': {'duration':500}}]
                     )  
                 ]
            )
        )
    ]
)

# The result
fig.show()
```
