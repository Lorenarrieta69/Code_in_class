```python
# Libraries
import pandas as pd
from sodapy import Socrata # it is the API
```

```python
# Define the source
client = Socrata("www.datos.gov.co", None)
```

```python
# Make the call 
r = client.get("32sa-8pi3", limit = 8000)
dr = pd.DataFrame.from_records(r)
```
